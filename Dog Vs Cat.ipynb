```python
import os
from keras.models import Sequential
from keras.layers import Conv2D , Dense , Flatten , MaxPooling2D
import cv2 as cv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
categories=os.listdir(os.path.join(os.getcwd(),'images'))
print(categories)
```

```python
def process_image(image_name,class_name):
    image = cv.imread(image_name,cv.IMREAD_GRAYSCALE)
    image= cv.reshape(image,dsize=(50,50))
    X.append(image)
    if(class_name=="Dog"):
        y.append(1)
    else: 
        y.append(0)
    return X,y
```

```python
X=[]
y=[]
for i in categories:
    image_lis=os.listdir(os.path.join(os.getcwd(),'images/'+str(i)))
    for j in image_lis:
        try:
            image = cv.imread((os.path.join(os.getcwd(),'images/'+str(i)+"/"+str(j))),cv.IMREAD_GRAYSCALE)
            image= cv.resize(image,dsize=(50,50))
            X.append(image)
            if(i=="Dog"):
                y.append(1)
            else:
                y.append(0)
            print((os.path.join(os.getcwd(),'images/'+str(i)+"/"+str(j))))
        except Exception as e:
            print(e)
            print("Some Problem With Image:" + str((os.path.join(os.getcwd(),'images/'+str(i)+"/"+str(j)))))
```

```python
import pickle 
X = np.array(X).reshape(-1, 50,50,1)
y = np.array(y)
pickle_out = open("X.pickle","wb")
pickle.dump(X, pickle_out)
pickle_out.close()
pickle_out = open("y.pickle","wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```

```python
X=X/255.0
```

```python
model = Sequential()
model.add(Conv2D(100,(5,5), activation = 'relu', input_shape = X.shape[1:]))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(Conv2D(200,(3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer="adam",
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=10, batch_size=32, validation_split=0.2)
```

```python
def preparing_test(image_name):
    image_test = cv.imread(image_name,cv.IMREAD_GRAYSCALE)
    image_test = cv.resize(image_test,dsize=(50,50))
    tt=[]
    tt.append(image_test)
    tt= np.array(tt).reshape(-1,50,50,1)
    tt = tt/255.0
    return tt
model.save('Dog Vs Cat V1.0.h5') # Saving Model predicted_val
```

```python
image = cv.imread('test.jpeg')
plt.imshow(image)
plt.show()
predictions = model.predict(preparing_test("test.jpeg"))
predicted_val = [int(round(p[0])) for p in predictions]
for j in predicted_val:
    if j==1:
        print("Dog")
    else:
        print("Cat")
```
